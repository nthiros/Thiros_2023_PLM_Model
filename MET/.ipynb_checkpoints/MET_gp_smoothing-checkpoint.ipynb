```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams['font.size']=12
```

```python
# Read in the MET forcing data for different water years
# combine into single array

wy = [15]#, 16, 17, 18, 19]
met_ = [np.loadtxt('wy{}_forcing.txt'.format(i)) for i in wy]
met = np.vstack(met_)
```

```python
# Move to pandas
tstart = pd.to_datetime('20{}-10-01 00'.format(min(wy)-1), format='%Y-%m-%d %H')
tend = pd.to_datetime('20{}-09-30 23'.format(max(wy)), format='%Y-%m-%d %H')

hours = pd.Series(pd.date_range(tstart, tend, freq='H'))
met_hr = pd.DataFrame(data=met, index=hours, columns=['rad_s','rad_l','prcp','temp','wnd_u','wnd_v','press','vap'])

units = {'rad_s':'Short Wave Radiation [W/m^2]',
         'rad_l':'Long Wave Radiation [W/m^2]',
         'prcp':'Precipitation [mm/s]',
         'temp':'Air Temperature [K]',
         'wnd_u':'East-to-West Wind [m/s]',
         'wnd_v':'South-to-North Wind [m/s]',
         'press':'Atmospheric Pressure [pa]',
         'vap':'Water-vapor Specific Humidity [kg/kg]'}

met_hr.head()
```

```python
# Use built-in pandas utilities to resmaple timeseries?
# Upscaling Rules
#   sum for prcp, average for all other variables

print (met_hr.shape)

met_day = met_hr.resample('1D').agg(dict(rad_s='mean',rad_l='mean',prcp='mean',temp='mean',wnd_u='mean',wnd_v='mean',press='mean',vap='mean'))
print (met_day.shape)
met_day.head()

met_month = met_hr.resample('1M').agg(dict(rad_s='mean',rad_l='mean',prcp='mean',temp='mean',wnd_u='mean',wnd_v='mean',press='mean',vap='mean'))
print (met_month.shape)
met_month.head()
```

```python
# Some plots
var = 'temp'
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(met_hr.index, met_hr[var], marker='None', alpha=0.4, label='hourly')
ax.plot(met_day.index, met_day[var], marker='None', label='day')
#ax.plot(met_month.index, met_month[var], marker='None', label='month')
ax.set_xlabel('Date')
ax.set_ylabel(units[var])
ax.legend()
plt.savefig('wy15_temp.png')
plt.show()
```

```python
var = 'prcp'
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(met_hr.index, met_hr[var], marker='None', alpha=0.4, label='hourly')
ax.plot(met_day.index, met_day[var], marker='None', label='day')
#ax.plot(met_month.index, met_month[var], marker='None', label='month')
ax.set_xlabel('Date')
ax.set_ylabel(units[var])
ax.legend()
plt.savefig('wy15_prcp.png')
plt.show()
```

```python
# Interpolate comparisions
# Upscale from months to days
#met_month2day = met_month.resample('D').interpolate('quadratic',limit_direction='both', pad=True)


# Downscale form days to hours
# Need to 'fill backwards' to not loose data
day_past = pd.DataFrame(data=met_day.iloc[-1,:]).T
day_past.index = [met_day.index[-1]+pd.Timedelta(1,'D')]

met_day_ = pd.concat((met_day,day_past))
met_day_.head()

met_day2hour = met_day_.resample('H').interpolate('quadratic',limit_direction='forward', pad=True)
met_day2hour = met_day2hour[:-1]
met_day2hour['prcp'][met_day2hour['prcp'] < 0.0] = 0.0

print (met_hr.shape)
print (met_day2hour.shape)

# Plots to compare the downscaling
var = 'prcp'
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(met_hr.index, met_hr[var], marker='None', alpha=0.4, label='hourly')
ax.plot(met_day2hour.index, met_day2hour[var], marker='None', label='day-2-hour')
ax.set_xlabel('Date')
ax.set_ylabel(units[var])
ax.legend()
plt.show()
```

```python
# Write some met files
fmt = '%.3f', '%.2f', '%.2E', '%.2f', '%.2f', '%.2f', '%.1f', '%.5f'

# Daily resolution 
np.savetxt('wy15_forcing_day2hr.txt', met_day2hour.to_numpy(), fmt=fmt, delimiter='\t')

# Hourly resolution, using daily signal though.
np.savetxt('wy15_forcing_daily.txt', met_day.to_numpy(), fmt=fmt, delimiter='\t')
```

```python
#---------------------------------------
# Make a new forcing dataframe
# Monthly resolution from 1979 to 2010
# Daily resolution from 2010 to 2021
#---------------------------------------
```

```python
# Read in the monthly NLDAS-2 forcing
monthly_full = pd.read_csv('./met_monthly_1979-2021.csv', index_col=0, parse_dates=True)
monthly_full.columns = ['rad_s','rad_l','prcp','temp','wnd_u','wnd_v','press','vap']
monthly_full.head()
```

```python
# Compare actual NLDAS monthly with hourly, downscaled to monthly
clip1 = pd.to_datetime('2014-10-01')
clip2 = pd.to_datetime('2018-09-30')

d1 = met_month[clip1:clip2]
d2 = monthly_full[clip1:clip2]

# Plots to compare the downscaling
var = 'prcp'
fig, ax = plt.subplots(figsize=(16,8))
ax.plot(d1.index, d1[var], marker='.', label='NLDAS_hr2month')
ax.plot(d2.index, d2[var], marker='.', label='NLDAS_monthly')
ax.legend()
plt.show()
```

```python
# Monthly to daily timeseries
# clip the months
mth_clip1 = pd.to_datetime('1979-10-01')
mth_clip2 = pd.to_datetime('2014-09-30')

day_clip1 = pd.to_datetime('2014-10-01')
day_clip2 = pd.to_datetime('2019-09-30')

month_clip = monthly_full[mth_clip1:mth_clip2]
day_clip   = met_day[day_clip1:day_clip2]


var = 'prcp'
ax.plot(day_clip.index, day_clip[var], marker='.', label='Daily')
ax.plot(month_clip.index, month_clip[var], marker='.', label='Monthly')
ax.legend()
plt.show()
```

```python
# is the mm of water per month preserved, regardless of timing?
m1 = month_clip[month_clip.index.year==2014]['prcp'][-1] # sept, 2015 precip in mm/sec
m1_mm = m1 * 30*(24*60*60) # accumulate the precip rate for the full month of sept.
print ('Total precipitation for sept, 2014 = {} mm -- monthly timesteps'.format(m1_mm)) # seems way too high

# Now for daily timeseries
d1 = day_clip[day_clip.index.year==2014]['prcp'][:31] # october, 2015 precip timeseries at daily timesteps, mm/sec still
d1_mm = d1 * 1*(24*60*60)
print ('Total precipitation for oct, 2014 = {} mm -- daily timesteps'.format(d1_mm.sum()))
```

```python
# Write some MET files
fmt = '%.3f', '%.2f', '%.2E', '%.2f', '%.2f', '%.2f', '%.1f', '%.5f'

# These are the monthly forcing between 1979 and 2014 water years
np.savetxt('wy1979-2014_monthly.txt', month_clip.to_numpy(), fmt=fmt, delimiter='\t')
# Daily forcing between 2015 and 2019 water years
np.savetxt('wy2015-2019_daily.txt', day_clip.to_numpy(), fmt=fmt, delimiter='\t')
```
